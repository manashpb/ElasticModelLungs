```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import os
```

```python
############ LL Lobe volume variation #############
folder = '/home/manash/DATA/SIMONE/COLLAB/2024/LUNG_NEW_MODEL_FULL/ELASTIC MODEL/PARAMETRIC/OUTPUT_kvar'
#folder = '/home/manash/DATA/SIMONE/COLLAB/2024/LUNG_NEW_MODEL_FULL/ELASTIC MODEL/PARAMETRIC/OUTPUT_damping'

fig, axes = plt.subplots(2, 3, figsize=(18, 10))
axes = axes.flatten()
lobesFiles = ['LL','LU','RL','RU','RM']

for i in range(5):
    path = os.path.join(folder,f'0.01/metric{lobesFiles[i]}.csv')
    #path = os.path.join(folder,f'1.0/metric{lobesFiles[i]}.csv')

    df = pd.read_csv(path, header=None)
    axes[i].plot(df[0],df[1], label=f'{lobesFiles[i]}')
    axes[i].hlines(df[1].mean(),0,df[0].max(),ls='--')
    axes[i].set_xlabel('Time')
    axes[i].set_ylabel('Volume')
    axes[i].set_title(f'{lobesFiles[i]} - Vol fluctuation from mean : {np.round((((df[1].max() - df[1].mean()) /  df[1].mean())*100),2)} %')
axes[5].axis('off')
```
